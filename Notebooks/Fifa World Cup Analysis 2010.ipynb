```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Fifa World Cup 2010 Datasets

```python
data = 'D:\\DML\\Fifa World Cup Datasets\\FIFA - 2010.csv'
fifa_data = pd.read_csv(data)
```

```python
top_10_teams = fifa_data.iloc[:10]
top_10_teams.head()
```

```python
fig = plt.figure(figsize=(14,8))
plt.title("Top 10 Teams By Points",size = 25, color="red")
sns.barplot(x = top_10_teams.Team, y = top_10_teams.Points,palette = "Set2")
```

```python
most_loss_teams = fifa_data.sort_values(by='Loss',ascending=False)
most_loss_teams.head(10)
```

```python
fig = plt.figure(figsize=(14,8))
top_5_loss_teams = most_loss_teams.head(5)
plt.title("Top 5 Teams By loss",size=15, color="red")
palette_color = sns.color_palette('dark')
explode = [0.2,0,0,0,0]
plt.pie(top_5_loss_teams.Loss,colors=palette_color,startangle = 150,shadow=True,labels = top_5_loss_teams.Team, explode=explode)
plt.legend()
plt.show()
```

```python
most_win = fifa_data.sort_values(by="Win", ascending=False)
most_win.head(10)
```

```python
fig = plt.figure(figsize= (14,8))
top_10_most_win_team = most_win.iloc[:10]
sns.barplot(x=top_10_most_win_team.Team, y=top_10_most_win_team.Win)
plt.title("Top 10 Teams by Wins", size=25)
```
